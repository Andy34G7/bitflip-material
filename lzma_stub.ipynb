```python
def lzma_compress(buffer_size:int = 4 * 1024 * 82, depth:int = 0):
    import lzma
    # Learn more: https://docs.python.org/3/library/lzma.html

    if buffer_size <= 4 * 1024:
        raise Exception("buffer_size must be atleast 4KiB")
    
    if buffer_size > 4  *  1024 ** 3:
        raise Exception("buffer_size must be at-most 4GiB")

    filters = [{
        "id": lzma.FILTER_LZMA1,
        "dict_size": buffer_size,
        "lc": 3,
        "lp": 0,
        "pb": 2,
        "mode": lzma.MODE_NORMAL,
        "mf": lzma.MF_BT4,
        "depth": depth
        }]

    with open("input.txt") as f:
        data = f.read().encode('utf-8')
        compressor = lzma.LZMACompressor(format=lzma.FORMAT_ALONE,filters=filters)
        compressed = compressor.compress(data)
        # Some data may still not be compressed, therefore we need to concatnate it
        compressed += compressor.flush()
        return compressed


print(lzma_compress())

```
